```python
#Diperoleh nilai 0 disetiap data NA maupun Null, artinya tidak ada missing value
print('Cek Value = 0 pada tenure')
print((df['tenure'].values == 0).sum())
print('Cek Value = 0 pada MonthlyCharges')
print((df['MonthlyCharges'].values == 0).sum())
```

```python
#Cek jumlah NaN dan Null pada TotalChargers
print('Jumlah NaN TotalCharges')
print(df['TotalCharges'].isna().sum())
print('Jumlah Null TotalCharges')
print(df['TotalCharges'].isnull().sum())
```

```python
#NaN dan Null bernilai 0, cek empty value
print('Empty Value pada TotalCharges')
print((df['TotalCharges'].values == ' ').sum())
```

```python
#ditemukan empty value, replace dengan NaN
df = df.replace(' ', np.nan)
df['TotalCharges'].isna().sum()
```

```python
#replace NaN dengan 0
df=df.fillna(value=0)
df['TotalCharges'].isna().sum()
```

```python
#Categorical Data Encoding
df['TotalCharges'] = df['TotalCharges'].astype('float64')
df.dtypes
```

```python
df=df.drop(columns = "customerID")
df.head()
```

```python
df_dummy=pd.get_dummies(df)
df_dummy
```

```python
#Anomalies and Outliers
import matplotlib.pyplot as plt
plt.boxplot(df['tenure'])
plt.show()
```

```python
Q1 = df['tenure'].quantile(0.25)
Q3 = df['tenure'].quantile(0.75)
IQR = Q3 - Q1
LB = Q1 - 1.5*IQR
UB = Q3 + 1.5*IQR
```

```python
Outliers_ten_UB = df[df['tenure']>UB]
Outliers_ten_UB
```

```python
Outliers_ten_LB = df[df['tenure']<LB]
Outliers_ten_LB
```

```python
Outliers_MC_UB = df[df['MonthlyCharges']>UB_MC]
Outliers_MC_UB
```

```python
Outliers_MC_LB = df[df['MonthlyCharges']<LB_MC]
Outliers_MC_LB
```
